```python
import torch
import matplotlib.pyplot as plt

from pathlib import Path
```

```python
path = Path("../results/").glob("**/*.pt")
results = [torch.load(x) for x in path]
```

```python
def plot_blobs(save):
    """2D plot of our blob data
    
    Args:
        labels: cluster assignment labels
    """
    fig, ax = plt.subplots(figsize=(20, 10))
    
    for res in save:
        data = res["data"]
        centroids = res["centroids"]
        labels = res["predictions"]
        
        ax.scatter(data[:, 0], data[:, 1], c=labels, s=100)

        if centroids is not None:
            ax.scatter(centroids[:, 0], centroids[:, 1], c='black', s=300, alpha=0.5);
    
    _ = ax.set_title('Blobs', fontsize=16)
```

```python
plot_blobs(results)
```
